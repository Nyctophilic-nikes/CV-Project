```python
import matplotlib.pyplot as plt
import numpy as np

# Data
datasets = ['MacaqueFaces', 'CTai', 'Czoo', 'NDD20v2', 'NyalaData', 'ZindiTurtleRecall']
WS = [75.72, 33.87, 67.61, 17.14, 10.75, 17.91]
WMD = [98.73, 91.10, 99.05, 67.42, 36.45, 74.40]
S = [74, 38, 7, 65, None, 25]  # Note: 'NyalaData' has missing data, represented as None
MD = [100, 96, 100, 70, 11, 22]

# Remove entries with missing data
datasets = [d for d, s in zip(datasets, S) if s is not None]
WS = [w for w, s in zip(WS, S) if s is not None]
WMD = [w for w, s in zip(WMD, S) if s is not None]
MD = [m for m, s in zip(MD, S) if s is not None]
S = [s for s in S if s is not None]

# Plot
fig, ax = plt.subplots(figsize=(10, 6))

# Using index as x-axis for better visualization
x = np.arange(len(datasets))

bar_width = 0.2

bar1 = plt.bar(x, WS, bar_width, label='WS')
bar2 = plt.bar(x + bar_width, WMD, bar_width, label='WMD')
bar3 = plt.bar(x + 2*bar_width, S, bar_width, label='S')
bar4 = plt.bar(x + 3*bar_width, MD, bar_width, label='MD')

# Set labels and title
plt.xlabel('Datasets')
plt.ylabel('Accuracy ')
plt.title('Accuracy by Dataset')
plt.xticks(x + 1.5*bar_width, datasets, rotation=45)
plt.legend()

# Display the plot
plt.tight_layout()
plt.show()
```
